```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\Administrator\Desktop\Exploratory Data Analysis\Daily Gold Price EDA\gold-historical-data-eda\dataset\gold.csv')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Close'], kde=True, color='gold')
plt.title('Distribution of Gold Prices')
plt.xlabel('Close')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Close', color='gold')
plt.title('Gold Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Year'] = df['Date'].dt.year

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Year', y='Close')
plt.title('Gold Price by Year')
plt.xlabel('Year')
plt.ylabel('Close')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Month'] = df['Date'].dt.month

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Month', y='Close')
plt.title('Gold Price by Month')
plt.xlabel('Month')
plt.ylabel('Close')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlOrBr')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
